```python
import pandas as pd
import numpy as np
import seaborn as sns

import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('dl_blverlauf.csv', sep=';')
```

```python
df.columns
```

```python
import matplotlib.ticker as ticker
sns.set(rc={'figure.figsize':(20,10)})
sns.set(font_scale = 1.8)
g = sns.lineplot(data = df,
            x='Datum',
            y='Genkopien pro EW / Tag * 10^6',
            hue = 'Gruppe')
g.xaxis.set_major_locator(ticker.LinearLocator(10))
g.set(title='Abwassermonitor Data Ã–sterreich')
```

```python
pv = df.pivot(index='Datum', columns = 'Gruppe', values = 'Genkopien pro EW / Tag * 10^6').head(-3)
```
